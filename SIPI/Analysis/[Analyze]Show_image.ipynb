```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load the image
# image_path = "page_87_200dpi.png"
image_path = "page87_cropped.png"
image = cv2.imread(image_path)

# Convert BGR to RGB for matplotlib display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Create figure and axis
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
ax.imshow(image_rgb)

# Define coordinates and labels
# coordinates = {
# "PVSIM" : (640, 540),  # LOWER
# "P3V3DS" : (230, 420),  # LEFT
# "KEY B TYPE WWAN_CONFIG_3" : (430, 210),  # LEFT
# "WWAN_TRANSMIT_OFF#" : (730, 260),  # RIGHT
# }

coordinates = {
  "PVSIM": (1100, 100),
  "KEY B TYPE": (1150, 600),
  "WWAN_CONFIG_3": (180, 130),
  "WWAN_TRANSMIT_OFF#": (1250, 130),
}



# Square size
square_size = 20

# Draw squares at each coordinate
for label, (x, y) in coordinates.items():
    # Create a rectangle (square) patch
    rect = patches.Rectangle(
        (x - square_size//2, y - square_size//2), 
        square_size, 
        square_size,
        linewidth=2,
        edgecolor='red',
        facecolor='none'
    )
    
    # Add the rectangle to the plot
    ax.add_patch(rect)
    
    # Add text label
    ax.text(x + square_size, y, label, 
           fontsize=8, 
           color='red', 
           fontweight='bold',
           bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

# Set title
ax.set_title("Image with Coordinate Squares")

# Remove axis ticks for cleaner display
ax.set_xticks([])
ax.set_yticks([])

# Show the image
plt.tight_layout()
plt.show()
```
